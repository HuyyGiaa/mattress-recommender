- Nệm Hybrid (Foam) -> Foam
- Nệm Hybrid (Cao su thiên nhiên + Lò xo túi độc lập + Foam),  Nệm Hybrid (Foam + Lò xo túi độc lập + Cao su thiên nhiên)
-  Nệm Hybrid (Cao su + Lò xo túi độc lập + Foam), Nệm Hybrid (Lò xo túi độc lập + cao su + Foam)

```python
mattress_dataset['category'] = mattress_dataset["material_type"].apply(extract_category)
```

```python
rows = []
for _, product in mattress_dataset.iterrows():
    base = {
        "product_name":  product.get("product_name"),
        "brand":         product.get("brand"),
        "material_type": product.get("material_type"),
        "category":      product.get("category"),
        "description":   product.get("description"),
        "rating":        product.get("rating"),
        "firmness":      product.get("specifications", {}).get("firmness") 
                         if isinstance(product.get("specifications"), dict) else None,
    }
    for var in product.get("variations", []):
        row = base.copy()
        row["size"]      = var.get("size")
        row["thickness"] = var.get("thickness")
        row["price"]     = var.get("price")
        rows.append(row)

df_flat = pd.DataFrame(rows)
print(df_flat.shape)
print(df_flat['price'].describe())
```
