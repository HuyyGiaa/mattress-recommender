## Chuẩn hoá các cột số và Firmness

### 1. Parse Size → Width & Length

**Vấn đề:** Cột `size` lưu dạng string `"160x200"`, `"160 x 200"` 
→ Không thể dùng trực tiếp cho model

**Xử lý:** Tách thành 2 cột số riêng biệt
- `"160x200"` → `width=160.0`, `length=200.0`
- Không parse được → `None`

---

### 2. Parse Thickness

**Vấn đề:** Cột `thickness` lưu dạng string `"10cm"`, `"10"`, `"10.5cm"`

**Xử lý:** Trích xuất số, bỏ đơn vị
- `"10cm"` → `10.0`
- `"10.5"` → `10.5`
- Không parse được → `None`

---

### 3. Chuẩn hoá Firmness

**Vấn đề:** Cột `firmness` có 14 giá trị khác nhau do:
- Viết hoa/thường không nhất quán: `"Trung bình"` vs `"Trung Bình"`
- Diễn đạt khác nhau cùng nghĩa: `"Cứng (Vững chắc)"` = `"Cứng vừa"`
- Description bị crawl nhầm vào firmness (chuỗi > 50 ký tự)
- Giá trị mơ hồ: `"Cao"` → map về `"Cứng"`

**Mapping chuẩn hoá về 7 nhóm:**

| Giá trị gốc | Sau chuẩn hoá |
|-------------|---------------|
| Trung bình, Trung Bình, Trung bình (Vững), Trung bình (Cân bằng), Độ cứng trung bình | `Trung bình` |
| Cứng vừa, Cứng trung bình, Cứng (Vững chắc) | `Cứng vừa` |
| Trung bình - Hơi cứng | `Trung bình - Hơi cứng` |
| Mềm - Trung bình, Mềm vừa, Mềm trung bình (Êm ái) | `Mềm - Trung bình` |
| Cứng, Cao | `Cứng` |
| Mềm | `Mềm` |
| Chuỗi > 50 ký tự (description nhầm) | `None` |

**Kết quả:** Từ 14 giá trị → 7 nhóm chuẩn
